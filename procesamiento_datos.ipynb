## D. Preparación de los datos

**Variables no utilizadas:**
* Se decidió no utilizar las variables **salary** y **salary_currency** ya que es necesario que la variable objetivo (salary_in_usd) se encuentre en las mismas unidades y estas dos variables no proporcionan nueva información.

**Variables utilizadas:**
* Se decidió utilizar las variables **work_year**,**experience_level**, **employment_type**, **job_title**, **employee_residence**, **remote_ratio**,**company_location**, **company_size** debido a que, como se observa en el punto anterior, tienen cierta influencia en el salario.

**Variable objetivo**
* Salario en dolares (salary_in_usd)

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
ordinal = pd.DataFrame()
encoder.fit(data[["experience_level","company_size"]])
ordinal[["experience_level","company_size"]] = encoder.transform(data[["experience_level","company_size"]])
ordinal['remote_ratio'] = data["remote_ratio"]
type(ordinal)
```

**Variables ordinales**

* Remote ratio
* Experience level
* Company size

**Variables nominales**

* work_year
* employment_type
* job_title
* employee_residence
* company_location

```python
work_year = pd.get_dummies(data.work_year, prefix='year')
employment_type = pd.get_dummies(data.employment_type, prefix='employment')
job_title = pd.get_dummies(data.job_title, prefix='job_title')
employee_residence = pd.get_dummies(data.employee_residence, prefix='employee_residence')
company_location = pd.get_dummies(data.company_location, prefix='company_location')

X = pd.concat([ordinal, work_year, employment_type, job_title, employee_residence, company_location], axis=1)
X
```
